```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# File paths
file_path_random_handy = "experiment_results_averaged/knowledge-gen-handy-time-final-random-average.csv"
file_path_random_trendy = "experiment_results_averaged/knowledge-gen-trendy-time-final-random-average.csv"
file_path_random_tweety = "experiment_results_averaged/knowledge-gen-tweety-time-final-random-average.csv"
file_path_random_jumpy = "experiment_results_averaged/knowledge-gen-jumpy-time-final-random-average.csv"
file_path_random_no_config = "experiment_results_averaged/knowledge-gen-no-config-time-final-random-average.csv"

# Load data
df_handy = pd.read_csv(file_path_random_handy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_trendy = pd.read_csv(file_path_random_trendy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_tweety = pd.read_csv(file_path_random_tweety, header=None, names=['Statement Count', 'Rank', 'Time'])
df_jumpy = pd.read_csv(file_path_random_jumpy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_no_config = pd.read_csv(file_path_random_no_config, header=None, names=['Statement Count', 'Rank', 'Time'])

# Pivot data
pivot_df_handy = df_handy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_trendy = df_trendy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_tweety = df_tweety.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_jumpy = df_jumpy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_no_config = df_no_config.pivot(index='Statement Count', columns='Rank', values='Time')

# Extract data for statement counts 350 and 300
statement_count_350_handy = pivot_df_handy.loc[350]
statement_count_350_trendy = pivot_df_trendy.loc[350]
statement_count_350_tweety = pivot_df_tweety.loc[350]
statement_count_350_jumpy = pivot_df_jumpy.loc[350]
statement_count_350_no_config = pivot_df_no_config.loc[350]

statement_count_300_handy = pivot_df_handy.loc[300]
statement_count_300_trendy = pivot_df_trendy.loc[300]
statement_count_300_tweety = pivot_df_tweety.loc[300]
statement_count_300_jumpy = pivot_df_jumpy.loc[300]
statement_count_300_no_config = pivot_df_no_config.loc[300]

# Define consistent colors
colors = sns.color_palette("deep", 5)

# Plotting
plt.figure(figsize=(10, 6))

# Solid lines for 350 statements
plt.plot(statement_count_350_handy.index, statement_count_350_handy.values, marker='o', linestyle='-', label='Handy (350)', color=colors[0])
plt.plot(statement_count_350_trendy.index, statement_count_350_trendy.values, marker='o', linestyle='-', label='Trendy (350)', color=colors[1])
plt.plot(statement_count_350_tweety.index, statement_count_350_tweety.values, marker='o', linestyle='-', label='Tweety (350)', color=colors[2])
plt.plot(statement_count_350_jumpy.index, statement_count_350_jumpy.values, marker='o', linestyle='-', label='Jumpy (350)', color=colors[3])
plt.plot(statement_count_350_no_config.index, statement_count_350_no_config.values, marker='o', linestyle='-', label='No Config (350)', color=colors[4])

# Dotted lines for 300 statements
plt.plot(statement_count_300_handy.index, statement_count_300_handy.values, marker='o', linestyle=':', label='Handy (300)', color=colors[0])
plt.plot(statement_count_300_trendy.index, statement_count_300_trendy.values, marker='o', linestyle=':', label='Trendy (300)', color=colors[1])
plt.plot(statement_count_300_tweety.index, statement_count_300_tweety.values, marker='o', linestyle=':', label='Tweety (300)', color=colors[2])
plt.plot(statement_count_300_jumpy.index, statement_count_300_jumpy.values, marker='o', linestyle=':', label='Jumpy (300)', color=colors[3])
plt.plot(statement_count_300_no_config.index, statement_count_300_no_config.values, marker='o', linestyle=':', label='No Config (300)', color=colors[4])

# Y-axis in log scale
plt.yscale('log')

# Adding labels, title, and grid
plt.xlabel('Rank')
plt.ylabel('Time (seconds)')
plt.title('Generation time for 350 and 300 statements - Random distribution (Y-axis in Log Scale)')
plt.grid(True)

# Legend outside the plot
plt.legend(title='Configurations', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
