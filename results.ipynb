*Random Distributions based on 350 statements* 


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path_random_handy="knowledge-gen-handy-time-final-random.csv"
file_path_random_trendy="knowledge-gen-trendy-time-final-random.csv"
file_path_random_tweety="knowledge-gen-random-time-tweety-final.csv" 
file_path_random_jumpy="knowledge-gen-random-time-jumpy-final.csv"
file_path_random_no_config = "knowledge-gen-no-config-time-final-random.csv"

df_handy = pd.read_csv(file_path_random_handy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_trendy = pd.read_csv(file_path_random_trendy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_tweety = pd.read_csv(file_path_random_tweety, header=None, names=['Statement Count', 'Rank', 'Time'])
df_jumpy = pd.read_csv(file_path_random_jumpy, header=None, names=['Statement Count', 'Rank', 'Time'])
df_no_config = pd.read_csv(file_path_random_no_config, header=None, names=['Statement Count', 'Rank', 'Time'])

pivot_df_handy = df_handy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_trendy = df_trendy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_tweety = df_tweety.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_jumpy = df_jumpy.pivot(index='Statement Count', columns='Rank', values='Time')
pivot_df_no_config = df_no_config.pivot(index='Statement Count', columns='Rank', values='Time')

statement_count_350_handy = pivot_df_handy.loc[350]
statement_count_350_trendy = pivot_df_trendy.loc[350]
statement_count_350_tweety = pivot_df_tweety.loc[350]
statement_count_350_jumpy = pivot_df_jumpy.loc[350]
statement_count_350_no_config = pivot_df_no_config.loc[350]

# Plotting the ranks for Statement Count 200 from the first file with solid lines
plt.figure(figsize=(10, 6))
plt.plot(statement_count_350_handy.index, statement_count_350_handy.values, marker='o', linestyle='-', label='Handy')
plt.plot(statement_count_350_trendy.index, statement_count_350_trendy.values, marker='o', linestyle='-', label='Trendy')
plt.plot(statement_count_350_tweety.index, statement_count_350_tweety.values, marker='o', linestyle='-', label='Tweety')
plt.plot(statement_count_350_jumpy.index, statement_count_350_jumpy.values, marker='o', linestyle='-', label='Jumpy')
plt.plot(statement_count_350_no_config.index, statement_count_350_no_config.values, marker='o', linestyle='-', label='No configuration')

# Adding labels, title, and legend
plt.xlabel('Rank')
plt.ylabel('Time(seconds)')
plt.title('Generation time for 350 statements- Random distribution')
plt.grid(True)
plt.legend(title='Distribution')
plt.show()
```
