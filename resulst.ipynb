```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data from the CSV file
data = pd.read_csv('base-rank-time.csv', header=None, names=['statements', 'ranks', 'time'])

# Extract the columns into separate lists
statements = data['statements']
ranks = data['ranks']
time = data['time']

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(statements, ranks, time, c='r', marker='x')

ax.set_xlabel('Amount of Statements')
ax.set_ylabel('Amount of Ranks')
ax.set_zlabel('Time to Compute')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('knowledge-gen-uniform-time.csv', header=None, names=['statements', 'ranks', 'time'])

# Get the unique ranks

df_filtered = df[~df['ranks'].isin([1,2,3,4,5,6,7,8,9,17])]
# Get the unique ranks (excluding rank 3)
ranks = df_filtered['ranks'].unique()

# Create a single plot for all ranks
plt.figure(figsize=(10, 6))

for rank in ranks:
    subset = df[df['ranks'] == rank]
    plt.plot(subset['statements'], subset['time'], marker='o', linestyle='-', label=f'Rank {rank}')

plt.title('Statements vs. Time for Different Ranks')
plt.xlabel('Statements')
plt.ylabel('Time')
plt.legend()  # Show legend to distinguish ranks
plt.grid(True)
plt.savefig('combined_plot.png')  # Save the figure
plt.show()  # Display the figure

print("Combined graph created.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'knowledge-gen-linear-time.csv'  # Replace with your actual file path
df = pd.read_csv(file_path, header=None, names=['Statement', 'Rank', 'Time'])

# Get the unique statements
statements = df['Statement'].unique()

# Plotting each statement
for statement in statements:
    # Filter the data for the current statement
    statement_data = df[df['Statement'] == statement]
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(statement_data['Rank'], statement_data['Time'], marker='o', linestyle='-', label=f'Statement {statement}')
    plt.xlabel('Rank')
    plt.ylabel('Time')
    plt.title(f'Statement {statement}: Rank vs Time')
    plt.grid(True)
    plt.legend()
    
    # Save each plot as a PNG file
    plt.savefig(f'statement_{statement}.png')
    
    # Show the plot
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file without named columns
file_path = 'knowledge-gen-linear-time.csv'  # Replace with your actual file path
df = pd.read_csv(file_path, header=None, names=['Statement', 'Rank', 'Time'])

# Pivot the data to have Ranks as the index and Statements as columns
pivot_df = df.pivot(index='Rank', columns='Statement', values='Time')

# Handle missing values by averaging
pivot_df = pivot_df.apply(lambda x: x.fillna(x.mean()), axis=1)

# Plotting all statements on one graph
plt.figure(figsize=(10, 6))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='o', linestyle='-', label=f'Statement {column}')

# Adding labels, title, and legend
plt.xlabel('Rank')
plt.ylabel('Time')
plt.title('Rank vs Time for All Statements')
plt.grid(True)
plt.legend(title='Statements')

# Save and show the plot
plt.savefig('all_statements_plot.png')
plt.show()
```
