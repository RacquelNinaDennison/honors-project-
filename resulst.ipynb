```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data from the CSV file
data = pd.read_csv('base-rank-time.csv', header=None, names=['statements', 'ranks', 'time'])

# Extract the columns into separate lists
statements = data['statements']
ranks = data['ranks']
time = data['time']

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(statements, ranks, time, c='r', marker='x')

ax.set_xlabel('Amount of Statements')
ax.set_ylabel('Amount of Ranks')
ax.set_zlabel('Time to Compute')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('knowledge-gen-uniform-time.csv', header=None, names=['statements', 'ranks', 'time'])

# Get the unique ranks

df_filtered = df[~df['ranks'].isin([1,2,3,4,5,6,7,8,9,17])]
# Get the unique ranks (excluding rank 3)
ranks = df_filtered['ranks'].unique()

# Create a single plot for all ranks
plt.figure(figsize=(10, 6))

for rank in ranks:
    subset = df[df['ranks'] == rank]
    plt.plot(subset['statements'], subset['time'], marker='o', linestyle='-', label=f'Rank {rank}')

plt.title('Statements vs. Time for Different Ranks')
plt.xlabel('Statements')
plt.ylabel('Time')
plt.legend()  # Show legend to distinguish ranks
plt.grid(True)
plt.savefig('combined_plot.png')  # Save the figure
plt.show()  # Display the figure

print("Combined graph created.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the data from the CSV file
df = pd.read_csv('knowledge-gen-uniform-time.csv', header=None, names=['statements', 'ranks', 'time'])

# Filter data to include only 200 statements
df_filtered = df[df['statements'] == 200]

# Create a 2D plot
plt.figure(figsize=(30, 10))

# Plot the data
plt.plot(df_filtered['ranks'], df_filtered['time'], marker='o', linestyle='-', color='r', label='Time vs. Ranks')

# Set labels and title
plt.xlabel('Ranks')
plt.ylabel('Time')
plt.title('Time vs Ranks for 200 Statements')

# Set x-axis ticks to be the values of the ranks
plt.xticks(ticks=df_filtered['ranks'].unique())

# Format x-axis ticks to show integer values
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.grid(True)
plt.legend()

# Save and show plot
plt.savefig('time_vs_ranks_200_statements.png')
plt.show()

print("Plot created and saved as 'time_vs_ranks_200_statements.png'.")
```

Random and uniform time for 250 statements 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the data from the two CSV files
df_random = pd.read_csv('knowledge-gen-random-time.csv', header=None, names=['statements', 'ranks', 'time'])
df_uniform = pd.read_csv('knowledge-gen-uniform-time-250.csv', header=None, names=['statements', 'ranks', 'time'])

# Filter data to include only 250 statements
df_random_filtered = df_random[df_random['statements'] == 250]
df_uniform_filtered = df_uniform[df_uniform['statements'] == 250]

# Sort the filtered DataFrames by 'ranks' and 'time'
df_random_sorted = df_random_filtered.sort_values(by=['ranks', 'time'])
df_uniform_sorted = df_uniform_filtered.sort_values(by=['ranks', 'time'])

# Create a 2D plot
plt.figure(figsize=(30, 10))

# Plot the random data
plt.plot(df_random_sorted['ranks'], df_random_sorted['time'], marker='o', linestyle='-', color='r', label='Random Time vs. Ranks')

# Plot the uniform data
plt.plot(df_uniform_sorted['ranks'], df_uniform_sorted['time'], marker='o', linestyle='-', color='b', label='Uniform Time vs. Ranks')

# Set labels and title
plt.xlabel('Ranks')
plt.ylabel('Time (seconds)')
plt.title('Time vs Ranks for 250 Statements')

# Set x-axis ticks to be the values of the ranks
plt.xticks(ticks=df_random_sorted['ranks'].unique())

# Format x-axis ticks to show integer values
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.grid(True)
plt.legend()

# Save and show plot
plt.show()
```
